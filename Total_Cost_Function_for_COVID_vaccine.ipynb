```python

import matplotlib.pyplot as plt
import numpy as np
import math
 
x = np.arange(0, 100, 0.1)

y1 = - np.exp( -x/21 ) + 20.5
y2 = - np.exp( -x/20 ) + 19.8
y3 = - np.exp( -x/17 ) + 9.5
 
plt.title("Total Cost Function")
plt.xlabel("Number of Vaccine (million)")
plt.ylabel("Total Cost (100 million dollar)")
plt.plot(x,y1,label='Pfizer')
plt.plot(x,y2,label='AstraZeneca')
plt.plot(x,y3,label='Jessen')
plt.legend()
plt.grid()
plt.show()

 
```
